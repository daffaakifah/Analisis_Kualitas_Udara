### Cleaning Data

```python
# Menghapus nilai null dari setiap dataset
data_Aotizhongxin_clean = data_Aotizhongxin.dropna()
data_Changping_clean = data_Changping.dropna()
data_Dingling_clean = data_Dingling.dropna()
data_Dongsi_clean = data_Dongsi.dropna()
data_Guanyuan_clean = data_Guanyuan.dropna()
data_Gucheng_clean = data_Gucheng.dropna()
data_Huairou_clean = data_Huairou.dropna()
data_Nongzhanguan_clean = data_Nongzhanguan.dropna()
data_Shunyi_clean = data_Shunyi.dropna()
data_Tiantan_clean = data_Tiantan.dropna()
data_Wanliu_clean = data_Wanliu.dropna()
data_Wanshouxigong_clean = data_Wanshouxigong.dropna()
```

```python
# Menampilkan jumlah missing value setelah penanganan
print("Jumlah missing value setelah data dihandling:")
print("Aotizhongxin:", data_Aotizhongxin_clean.isna().sum().sum())
print("Changping:", data_Changping_clean.isna().sum().sum())
print("Dingling:", data_Dingling_clean.isna().sum().sum())
print("Dongsi:", data_Dongsi_clean.isna().sum().sum())
print("Guanyuan:", data_Guanyuan_clean.isna().sum().sum())
print("Gucheng:", data_Gucheng_clean.isna().sum().sum())
print("Huairou:", data_Huairou_clean.isna().sum().sum())
print("Nongzhanguan:", data_Nongzhanguan_clean.isna().sum().sum())
print("Shunyi:", data_Shunyi_clean.isna().sum().sum())
print("Tiantan:", data_Tiantan_clean.isna().sum().sum())
print("Wanliu:", data_Wanliu_clean.isna().sum().sum())
print("Wanshouxigong:", data_Wanshouxigong_clean.isna().sum().sum())
```

```python
# Menggabungkan semua dataset yang sudah dibersihkan
all_df_clean = pd.concat([data_Aotizhongxin_clean, data_Changping_clean, data_Dingling_clean, data_Dongsi_clean,
                          data_Guanyuan_clean, data_Gucheng_clean, data_Huairou_clean, data_Nongzhanguan_clean,
                          data_Shunyi_clean, data_Tiantan_clean, data_Wanliu_clean, data_Wanshouxigong_clean],
                         ignore_index=True)
```

```python
# Mengubah kolom 'date' menjadi tipe datetime
all_df_clean['date'] = pd.to_datetime(all_df_clean[['year', 'month', 'day', 'hour']])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

Bagaimana tren rata-rata tahunan dari berbagai polutan (PM2.5, PM10, SO2, NO2, CO, O3) serta polutan apa yang paling tinggi secara keseluruhan dan persebaran di masing-masing stasiun?

**Trend polutan per tahun**

```python
# Menghitung rata-rata tahunan dari berbagai polutan
annual_pollutant_means = all_df_clean.groupby(all_df_clean['date'].dt.year)[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean()
```

```python
# Plot tren rata-rata tahunan dari berbagai polutan
plt.figure(figsize=(14, 8))
plt.plot(annual_pollutant_means.index, annual_pollutant_means['PM2.5'], label='PM2.5', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['PM10'], label='PM10', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['SO2'], label='SO2', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['NO2'], label='NO2', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['CO'], label='CO', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['O3'], label='O3', marker='o')
plt.title('Tren Rata-rata Tahunan dari Berbagai Polutan')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Konsentrasi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Menentukan polutan dengan rata-rata tertinggi
highest_pollutant = annual_pollutant_means.mean().idxmax()
highest_pollutant_mean = annual_pollutant_means.mean().max()

print(f"Polutan dengan rata-rata tertinggi adalah {highest_pollutant} dengan rata-rata konsentrasi {highest_pollutant_mean:.2f}.")
```

```python
# Menentukan polutan dengan rata-rata tertinggi
highest_pollutant = annual_pollutant_means.mean().idxmax()
highest_pollutant_mean = annual_pollutant_means.mean().max()

print(f"Polutan dengan rata-rata tertinggi adalah {highest_pollutant} dengan rata-rata konsentrasi {highest_pollutant_mean:.2f}.")
```

**Insight**:
- Polutan dengan rata-rata tertinggi adalah CO dengan rata-rata konsentrasi 1287.94 dengan rata-rata kadarnya tertinggi pada tahun 2017 yaitu mencapai 1600
