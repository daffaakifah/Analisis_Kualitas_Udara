## Visualization & Explanatory Analysis

### Pertanyaan 1:

Bagaimana tren rata-rata tahunan dari berbagai polutan (PM2.5, PM10, SO2, NO2, CO, O3) serta polutan apa yang paling tinggi?

```python
# Mengubah kolom 'date' menjadi tipe datetime
all_df_clean['date'] = pd.to_datetime(all_df_clean[['year', 'month', 'day', 'hour']])
```

```python
# Menghitung rata-rata tahunan dari berbagai polutan
annual_pollutant_means = all_df_clean.groupby(all_df_clean['date'].dt.year)[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean()
```

```python
# Plot tren rata-rata tahunan dari berbagai polutan
plt.figure(figsize=(14, 8))
plt.plot(annual_pollutant_means.index, annual_pollutant_means['PM2.5'], label='PM2.5', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['PM10'], label='PM10', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['SO2'], label='SO2', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['NO2'], label='NO2', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['CO'], label='CO', marker='o')
plt.plot(annual_pollutant_means.index, annual_pollutant_means['O3'], label='O3', marker='o')
plt.title('Tren Rata-rata Tahunan dari Berbagai Polutan')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Konsentrasi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Menentukan polutan dengan rata-rata tertinggi
highest_pollutant = annual_pollutant_means.mean().idxmax()
highest_pollutant_mean = annual_pollutant_means.mean().max()

print(f"Polutan dengan rata-rata tertinggi adalah {highest_pollutant} dengan rata-rata konsentrasi {highest_pollutant_mean:.2f}.")
```

### Pertanyaan 2:

Bagaimana pengaruh suhu, curah hujan, kecepatan angin, dan tekanan udara terhadap konsentrasi CO?

```python
# Plot hubungan antara suhu dan CO
plt.figure(figsize=(14, 6))
plt.scatter(all_df_clean['TEMP'], all_df_clean['CO'], alpha=0.5)
plt.title('Hubungan antara Suhu dan CO')
plt.xlabel('Suhu (C)')
plt.ylabel('Konsentrasi CO')
plt.grid(True)
plt.show()
```

**Insight:**


```python
# Plot hubungan antara curah hujan dan CO
plt.figure(figsize=(14, 6))
plt.scatter(all_df_clean['RAIN'], all_df_clean['CO'], alpha=0.5)
plt.title('Hubungan antara Curah Hujan dan CO')
plt.xlabel('Curah Hujan (mm)')
plt.ylabel('Konsentrasi CO')
plt.grid(True)
plt.show()
```

```python
# Plot hubungan antara kecepatan angin dan CO
plt.figure(figsize=(14, 6))
plt.scatter(all_df_clean['WSPM'], all_df_clean['CO'], alpha=0.5)
plt.title('Hubungan antara Kecepatan Angin dan CO')
plt.xlabel('Kecepatan Angin (m/s)')
plt.ylabel('Konsentrasi CO')
plt.grid(True)
plt.show()
```

```python
# Plot hubungan antara tekanan udara dan CO
plt.figure(figsize=(14, 6))
plt.scatter(all_df_clean['PRES'], all_df_clean['CO'], alpha=0.5)
plt.title('Hubungan antara Tekanan Udara dan CO')
plt.xlabel('Tekanan Udara (hPa)')
plt.ylabel('Konsentrasi CO')
plt.grid(True)
plt.show()
```

## Analisis Lanjutan

```python
# Fungsi untuk mengkategorikan level CO
def kategorikan_co(co_value):
    if co_value < 70:
        return "Aman"
    elif 70 <= co_value < 150:
        return "Tinggi"
    elif co_value >= 150:
        return "Sangat tinggi"
    else:
        return "Tidak Diketahui"

# Menerapkan fungsi kategorisasi ke kolom CO
all_df_clean['Kategori_CO'] = all_df_clean['CO'].apply(kategorikan_co)

# Membuat tabel ringkasan jumlah kategori CO per stasiun
ringkasan_kategori_co = all_df_clean.groupby('station')['Kategori_CO'].value_counts().unstack(fill_value=0)

# Menampilkan tabel ringkasan
print(ringkasan_kategori_co)

# Menyimpan tabel ringkasan ke file CSV
ringkasan_kategori_co.to_csv("ringkasan_kategori_co.csv")
```

## Conclusion

- Kadar gas yang memiliki rata-rata tertinggi adalah gas CO
- Kecepatan angin dan curah hujan mempengaruhi kadar CO. Apabila kecepatan angin dan curah hujan tinggi maka dapat mengurangi konsentrasi kadar CO
- Stasiun kerap kali memiliki kadar CO yang sangat tinggi yang dapat berbahaya bagi kesehatan, dperlukan tindakan lebih lanjut untuk menangani kadar konsentrasi CO yang tinggi.
